```python
## Computational Foundations of Systems Biology
## Assignment 1 - Question 1
## Yamuna Rajan

import numpy as np
import matplotlib.pyplot as plt

N = 100                                              # total number of data points
n = np.arange(0, N, 1)                               # [0,..., N-1] (vector)

x = np.empty(N)                                      # protein x concentration in uM (vector)
x.fill(10)                                           # Protein x levels are constant (= 10 uM)

y = np.empty(N)                                      # protein y concentration in uM (vector)
z = np.empty(N)                                      # protein z concentration in uM (vector)

y[0] = 0                                             # initial y protein concentration (= 0 uM)
z[0] = 0                                             # initial z protein concentration (= 0 uM)

Pxy = 0.3                                            # Production parameter Pxy
Pxz = 0.3                                            # Production parameter Pxz
Pyz = 0.3                                            # Production parameter Pyz
Px = 0.9                                             # degradation parameter Px
Py = 0.9                                             # degradation parameter Py
Pz = 0.9                                             # degradation parameter Pz

for i in range (1, N):                               # discrete-time index i
    y[i] = Pxy*x[i-1] + Py*y[i-1]                    # difference equation for protein degradation
    z[i] = Pxz*x[i-1] + Pyz*y[i-1] + Pz*z[i-1]

plt.plot(n,x,'g',label = 'x')
plt.plot(n,y,'r',label = 'y')
plt.plot(n,z,'b',label = 'z')
plt.xlabel('time (i)')
plt.ylabel('protein concentration (uM)')
plt.legend(loc='upper right')
plt.title('3 gene network')
plt.grid(True)
plt.show()
```

```python
## Computational Foundations of Systems Biology
## Assignment 1 - Question 2 Part 2
## Yamuna Rajan

import numpy as np
import math
import matplotlib.pyplot as plt
from fractions import Fraction

x = np.array([6, 3, 9, 4])                              # protein levels at 4 distinct time points
N = len(x)                                              # total number of data points
X = np.zeros(N, dtype=complex)                          # a vector of type complex and size N intialized to 0 + 0j
magnitude = np.empty(N)                                 # an empty vector of size N to store the magnitude of X

def e_function(m, n):                                   # a function to calculate the value of e^(pi*m*n/2)
    theta = (math.pi/2)*m*n                             # value of theta
    return complex(math.cos(theta), math.sin(theta))    # using the identity e^theta = cos(theta) + i*sin(theta)

for i in range(N):
    for j in range(N):
        X[i] += x[j]*e_function(i, j)                   # DFT formula
        
for i in range(N):
    magnitude[i] = abs(X[i])                            # calculating the magnitude

omega = [0]                                             # a list store the values of (2*pi*m/N)
omega_ticks = [0]                                       # a list store the labels of omega
for i in range(1, N):
    omega.append((2*i)/N)
    omega_ticks.append(str(Fraction(omega[i]).numerator) + "pi/" +  str(Fraction(omega[i]).denominator))

for i in range(N):
    plt.text(omega[i]-0.03, magnitude[i]+1, str(round(magnitude[i],2)))
    
plt.xticks(omega, omega_ticks)
plt.bar(omega, magnitude, width=0.01)
plt.scatter(omega, magnitude)
plt.ylim([0, 25])
plt.xlabel('w(m)')
plt.ylabel('|X(w(m))|')
plt.title('Discrete Fourier Transform plot')
plt.grid(True)
plt.show()
```
